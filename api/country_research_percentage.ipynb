```python
import pandas as pd
```

```python
# Load the dataset
data = pd.read_csv("country_research_number.csv")  # Replace with your file name

# Calculate the total number of researches for each country and year
data['Total Researches'] = data.iloc[:, 2:].sum(axis=1)  # Sum across subject columns
```

```python
# Drop the 'Undefined' column if it exists
data = data.drop(columns=['Undefined'], errors='ignore')  # 'errors=ignore' prevents errors if the column doesn't exist
```

```python
# Convert counts to percentages
for column in data.columns[2:-1]:  # Exclude 'Country', 'Year', and 'Total Researches'
    data[column] = ((data[column] / data['Total Researches']) * 100).round(2)
```

```python
# Save the percentage data to a new CSV
data.to_csv("nawa_percent_2.csv", index=False)

# Display the resulting DataFrame
print(data.head())
```
